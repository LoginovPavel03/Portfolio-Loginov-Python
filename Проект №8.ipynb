### А/В-эксперимент

В эксперименте участвовали новые пользователи, которые зарегистрировались в приложении со 2 по 15 июня 2025 года. 
Пользователей случайным образом разделили на две равные группы: 
- Контрольная группа проходила стандартный онбординг без обязательного обучения.
- Тестовая группа проходила обновлённый онбординг с подробной информацией об активах и связанных с ними рисках. 

После разделения активность пользователей анализировали в течение недели.

Цель эксперимента — оценить, как обновлённый онбординг влияет на поведение пользователей и их дальнейшую инвестиционную активность.

При внедрении новой фичи команда продукта выдвинула такие гипотезы:
- Гипотеза роста: обучающий онбординг помогает пользователям лучше понимать принципы инвестирования, поэтому они будут чаще открывать второй депозит.
- Гипотеза риска: информация о возможных потерях и высоких рисках отпугнёт некоторых новичков, особенно самых осторожных, что снизит конверсию в первый депозит.
- Дополнительная гипотеза: после нового онбординга пользователи, которые выбрали высокорискованные активы, будут чаще, чем раньше, возвращаться и открывать второй депозит. При старом онбординге пользователи часто покупали активы с высоким риском без понимания последствий. Это приводило к потерям и оттоку после первого депозита. 

Чтобы увидеть полную картину, недостаточно следить за метриками конверсии. Команда выбрала для эксперимента такой набор метрик: 
- Ключевая метрика — средняя сумма всех депозитов на одного пользователя (включая тех, кто установил приложение или открыл веб-версию).
- Барьерная метрика — конверсия из регистрации в первый депозит.
- Вспомогательная метрика 1 — конверсия из первого депозита во второй.
- Вспомогательная метрика 2 — средняя сумма всех депозитов на пользователя, который открыл хотя бы один депозит.

Ожидалось, что ключевая и барьерная метрики не упадут, а вспомогательные покажут значительный рост.

**Задача 2.2.** Анализ воронок событий.

Построим и визуализируем общую воронку (по всем данным), затем опишем пользовательский путь: какие этапы представлены, какова их последовательность, на каком (или на каких) этапах теряется большая часть новых пользователей.
    
При решении построим две воронки: 
- **классическую воронку** — доля от первого шага в процентах
- **step-by-step** — с конверсией от предыдущего этапа.

Предполагается, что пользователь проходит такой путь к совершению второго депозита:
1. `install / open_web` — пользователь скачивает или открывает приложение в браузере
2. `introduction` — пользователь видит введение
2. `registration` — пользователь регистрируется в приложении
2. `main_page` — пользователь попадает на основную страницу (личный кабинет)
2. `onboarding_complete` — пользователь заканчивает проходить онбординг
2. `first_deposit` — пользователь совершает первый депозит
2. `asset_purchase` — пользователь приобретает актив
2. `second_deposit` — пользователь совершает второй депозит

Начнём с построения **классической** воронки, которая будет отражать долю от первого шага в процентах. Представим её в виде таблицы, куда также добавим **step-by-step** воронку для удобства анализа и построения визуализации

```python
# Задаем порядок шагов в воронке

funnel_steps = ['install / open_web', 'introduction', 'registration', 'main_page',  
                'onboarding_complete', 'first_deposit', 'asset_purchase', 'second_deposit']

# Фильтруем датасет оставляя только нужные шаги воронки

funnel_hist_steps = hist[hist['event_name'].isin(funnel_steps)]

# Считаем число уникальных пользователей на каждом шаге

funnel_counts = funnel_hist_steps.groupby('event_name')['user_id'].nunique().reindex(funnel_steps)

# Собираем таблицу воронки

funnel_hist = pd.DataFrame({
   'step': funnel_steps,
   'users': funnel_counts.values
})

# Строим классическую воронку

funnel_hist['conversion_from_first_%'] = (funnel_hist['users'] / funnel_hist['users'].iloc[0] * 100).round(1)

# Визуализируем классическую воронку

plt.figure(figsize=(10, 6))
plt.bar(funnel_hist['step'], funnel_hist['conversion_from_first_%'], color='blue')
plt.title('Классическая воронка: конверсия от первого шага')
plt.ylabel('Конверсия от первого шага (%)')
plt.xlabel('Этапы воронки')
plt.xticks(rotation = 45)
plt.show()

# Строим step-by-step воронку

funnel_hist['conversion_from_previous_%'] = (funnel_hist['users'] / funnel_hist['users'].shift(1) * 100).round(1)

# Визуализируем step-by-step воронку

plt.figure(figsize=(10, 6))
plt.bar(funnel_hist['step'][2:], funnel_hist['conversion_from_previous_%'][2:], color='skyblue')
plt.title('Step-by-step воронка: конверсия от предыдущего шага')
plt.ylabel('Конверсия от предыдущего шага (%)')
plt.xlabel('Этапы воронки')
plt.xticks(rotation = 45)
plt.show()

# Выводим таблицу воронки

print(funnel_hist)
```

После построения **классической** воронки можно увидеть, что конверсия шагов от первого шага изменяется стабильно, без резких снижений. Конверсия в целевое действие (`second_deposit`) составляет 13% - с данной метрикой стоит поработать для увеличения прибыли компании

После построения **step-by-step** воронки мы обратили внимание, что конверсия на этапах установки/открытия и введения в приложения - одинаковы, поэтому на графике данные этапы - не отображаются. Заметим, что на этапах приобретения актива (`asset_purchase`) и совершения второго депозита (`second_deposit`) происходит набольшая потеря пользователей (55% и 35% соответственно). Возможно, проблема возникает не во всех странах и не на всех типах устройств, а только некоторых

Для этого мы построим **step-by-step** воронку относительно предыдущего шага отдельно для каждой страны, а также для мобильной и web версий

```python
# Разделим данные по платформам и оставим только нужные шаги воронки

counts_br = hist[(hist['country_code'] == 'BR') & (hist['event_name'].isin(funnel_steps))]
counts_mx = hist[(hist['country_code'] == 'MX') & (hist['event_name'].isin(funnel_steps))]
counts_ar = hist[(hist['country_code'] == 'AR') & (hist['event_name'].isin(funnel_steps))]
counts_co = hist[(hist['country_code'] == 'CO') & (hist['event_name'].isin(funnel_steps))]
counts_mobile = hist[(hist['platform'] == 'mobile') & (hist['event_name'].isin(funnel_steps))]
counts_web = hist[(hist['platform'] == 'web') & (hist['event_name'].isin(funnel_steps))]

# Считаем количество пользователей на каждом шаге

counts_br_agg = counts_br.groupby('event_name')['user_id'].nunique().reindex(funnel_steps).reset_index()
counts_mx_agg = counts_mx.groupby('event_name')['user_id'].nunique().reindex(funnel_steps).reset_index()
counts_ar_agg = counts_ar.groupby('event_name')['user_id'].nunique().reindex(funnel_steps).reset_index()
counts_co_agg = counts_co.groupby('event_name')['user_id'].nunique().reindex(funnel_steps).reset_index()
counts_mobile_agg = counts_mobile.groupby('event_name')['user_id'].nunique().reindex(funnel_steps).reset_index()
counts_web_agg = counts_web.groupby('event_name')['user_id'].nunique().reindex(funnel_steps).reset_index()

# Считаем конверсию и округляем значение до 1 знака после запятой

br_funnel = counts_br_agg['user_id'] / counts_br_agg['user_id'].shift(1) * 100
mx_funnel = counts_mx_agg['user_id'] / counts_mx_agg['user_id'].shift(1) * 100
ar_funnel = counts_ar_agg['user_id'] / counts_ar_agg['user_id'].shift(1) * 100
co_funnel = counts_co_agg['user_id'] / counts_co_agg['user_id'].shift(1) * 100
mobile_funnel = counts_mobile_agg['user_id'] / counts_mobile_agg['user_id'].shift(1) * 100
web_funnel = counts_web_agg['user_id'] / counts_web_agg['user_id'].shift(1) * 100

# Соберём через словарь данные в датафрейм и исключим первый и второй шаги для каждого элемента

funnel_compare = pd.DataFrame({
    'step': funnel_steps[2:],
    'brazil': br_funnel[2:].round(1).values,
    'mexico': mx_funnel[2:].round(1).values,
    'argentina': ar_funnel[2:].round(1).values,
    'columbia': co_funnel[2:].round(1).values,
    'mobile': mobile_funnel[2:].round(1).values,
    'web': web_funnel[2:].round(1).values,
})

# Визуализируем воронку для Бразилии

plt.figure(figsize=(10, 6))
plt.bar(funnel_compare['step'], funnel_compare['brazil'], color='yellow')
plt.title('Step-by-step воронка для Бразилии: конверсия от предыдущего шага')
plt.ylabel('Конверсия от первого шага (%)')
plt.xlabel('Этапы воронки')
plt.show()

# Визуализируем воронку для Мексики

plt.figure(figsize=(10, 6))
plt.bar(funnel_compare['step'], funnel_compare['mexico'], color='orange')
plt.title('Step-by-step воронка для Мексики: конверсия от предыдущего шага')
plt.ylabel('Конверсия от первого шага (%)')
plt.xlabel('Этапы воронки')
plt.show()

# Визуализируем воронку для Аргентины

plt.figure(figsize=(10, 6))
plt.bar(funnel_compare['step'], funnel_compare['argentina'], color='lightblue')
plt.title('Step-by-step воронка для Аргентины: конверсия от предыдущего шага')
plt.ylabel('Конверсия от первого шага (%)')
plt.xlabel('Этапы воронки')
plt.show()

# Визуализируем воронку для Колумбии

plt.figure(figsize=(10, 6))
plt.bar(funnel_compare['step'], funnel_compare['columbia'], color='green')
plt.title('Step-by-step воронка для Колумбии: конверсия от предыдущего шага')
plt.ylabel('Конверсия от первого шага (%)')
plt.xlabel('Этапы воронки')
plt.show()

# Визуализируем mobile-версию

plt.figure(figsize=(10, 6))
plt.bar(funnel_compare['step'], funnel_compare['mobile'], color='blue')
plt.title('Step-by-step воронка для mobile-версии: конверсия от предыдущего шага')
plt.ylabel('Конверсия от предыдущего шага (%)')
plt.xlabel('Этапы воронки')
plt.show()

# Визуализируем web-версию

plt.figure(figsize=(10, 6))
plt.bar(funnel_compare['step'], funnel_compare['web'], color='red')
plt.title('Step-by-step воронка для web-версии: конверсия от предыдущего шага')
plt.ylabel('Конверсия от первого шага (%)')
plt.xlabel('Этапы воронки')
plt.show()

# Выводим таблицу воронки

print(funnel_compare)
```

После построения **step-by-step** воронки относительно предыдущего шага отдельно для всех стран, а также мобильной и web версии, мы видим, что показатели конверсии на каждом из этапов приблизительно равны в разрезе всех стран и типов устройств. А значит, что проблема основной потери аудитории не зависит от страны или типа устройства

**Задача 2.3.** Влияние уровня риска актива на открытие второго депозита.

Для того, чтобы понять, как уровень риска первого актива влияет на конверсию этапа совершения второго депозита, мы рассчитаем показатель конверсии этапа совершения второго депозита для каждого уровня риска первого приобрётённого актива

```python
# Фильтруем только события покупки

asset_df = hist[hist['event_name'] == 'asset_purchase']

# Определяем первый купленный актив по времени

first_purchase = asset_df.sort_values(by='event_ts').drop_duplicates(subset='user_id', keep='first')

# Оставляем нужные поля

first_purchase = first_purchase[['user_id', 'risk_level']]

# Пользователи, у которых был второй депозит

second_deposit_users = hist[hist['event_name'] == 'second_deposit']['user_id'].unique()

# Добавим флаг

first_purchase['has_second_deposit'] = first_purchase['user_id'].isin(second_deposit_users)

# Группируем по risk_level

conversion_by_risk = first_purchase.groupby('risk_level')['has_second_deposit']\
                                   .agg(['count', 'sum'])\
                                   .rename(columns={'count': 'users', 'sum': 'second_deposit_users'})

# Считаем конверсию    
    
conversion_by_risk['conversion_rate_%'] = (conversion_by_risk['second_deposit_users'] / conversion_by_risk['users'] * 100).round(2)
conversion_by_risk.reset_index(inplace=True)
conversion_by_risk.sort_values(by = 'conversion_rate_%', ascending = False)
```

После расчёта показателя конверсии этапа совершения второго депозита для каждого уровня риска первого приобретённого актива, мы видим, что с увеличением уровня риска первого приобретённого актива снижается конверсия этапа совершения второго депозита

**Задача 2.4.** Анализ ключевой метрики на исторических данных.

Рассчитаем ключевую метрику - среднюю сумму всех депозитов на одного пользователя. Будем учитывать всех пользователей, взаимодействующих с сервисом. Это необходимо для фиксации базового уровня монетизации

```python
# Группируем по пользователям, суммируем все их депозиты и усредняем сумму всех депозитов на одного пользователя

avg_deposit_per_user = hist.groupby('user_id')['amount'].sum().mean()
deposit_per_user_grouped = hist.groupby('user_id')['amount'].sum().reset_index()
deposit_avg_per_user_grouped = deposit_per_user_grouped.groupby('user_id')['amount'].mean().reset_index()
std_avg_deposit_per_user = deposit_avg_per_user_grouped.groupby('user_id')['amount'].sum().std()

print(f"Средняя сумма всех депозитов на одного пользователя: {avg_deposit_per_user:.2f}")
print(f"Стандартное отклонее суммы всех депозитов на одного пользователя: {std_avg_deposit_per_user:.2f}")
```

Получаем значение ключевой метрики - `88.47` денежных единиц, а также её стандартного отклонения - `79.64`

Дополнительно построим гистограмму распределения, которая поможет визуально определить перекос распределения, а также к ней добавим описательную статистику

```python
# Строим гистограмму распределения

plt.figure(figsize=(12, 8))
plt.hist(deposit_avg_per_user_grouped['amount'], bins=30, color = 'lightblue', edgecolor = 'black')
plt.title("Гистограмма распределения средней суммы всех депозитов на одного пользователя")
plt.ylabel('Конверсия от первого шага (%)')
plt.xlabel('Этапы воронки')
plt.show()

deposit_avg_per_user_grouped['amount'].describe(percentiles=[0.25, 0.5, 0.75, 0.8, 0.85, 0.9, 0.95, 0.99])
```

На основе построенной гистограммы распределения ключевой метрики, мы можем увидеть, что наблюдается правосторонняя асимметрия графика, то есть, данные распределены ненормально. **Задача 3.2.** Сравнение воронок событий.

Построим и визуализируем общую воронку (по всем данным), затем опишем пользовательский путь: какие этапы представлены, какова их последовательность, на каком (или на каких) этапах теряется большая часть новых пользователей.
    
При решении построим две воронки: 
- **классическую воронку** — доля от первого шага в процентах
- **step-by-step** — с конверсией от предыдущего этапа.

Затем визуализируем и сравним эти воронки с помощью линейных графиков

```python
# Задаем порядок шагов в воронке

funnel_steps = ['install / open_web', 'introduction', 'registration', 'main_page',  
                'onboarding_complete', 'first_deposit', 'asset_purchase', 'second_deposit']

# Фильтруем датасет оставляя только нужные шаги воронки

funnel_abt_steps = abt[abt['event_name'].isin(funnel_steps)]

# Фильтруем датасет, оставляя данные контрольной группы

control_funnel_abt_steps = funnel_abt_steps[funnel_abt_steps['group'] == 'control']

# Фильтруем датасет, оставляя данные тестовой группы

test_funnel_abt_steps = funnel_abt_steps[funnel_abt_steps['group'] == 'test']

# Считаем число уникальных пользователей на каждом шаге

control_funnel_counts = control_funnel_abt_steps.groupby('event_name')['user_id'].nunique().reindex(funnel_steps)
test_funnel_counts = test_funnel_abt_steps.groupby('event_name')['user_id'].nunique().reindex(funnel_steps)

# Собираем таблицу воронки

funnel_abt = pd.DataFrame({
   'step': funnel_steps,
   'users_control': control_funnel_counts.values,
   'users_test': test_funnel_counts.values
})

# Строим классическую воронку

funnel_abt['conversion_from_first_%_control'] = (funnel_abt['users_control'] / funnel_abt['users_control'].iloc[0] * 100).round(1)
funnel_abt['conversion_from_first_%_test'] = (funnel_abt['users_test'] / funnel_abt['users_test'].iloc[0] * 100).round(1)

# Визуализируем классическую воронку

plt.figure(figsize=(10, 6))
plt.plot(funnel_abt['step'], funnel_abt['conversion_from_first_%_control'], color='blue', label = 'control')
plt.plot(funnel_abt['step'], funnel_abt['conversion_from_first_%_test'], color='red', label = 'test')
plt.title('Классическая воронка: конверсия от первого шага')
plt.ylabel('Конверсия от первого шага (%)')
plt.xlabel('Этапы воронки')
plt.xticks(rotation = 45)
plt.grid(True)
plt.legend()
plt.show()

# Строим step-by-step воронку

funnel_abt['conversion_from_previous_%_control'] = (funnel_abt['users_control'] / funnel_abt['users_control'].shift(1) * 100).round(1)
funnel_abt['conversion_from_previous_%_test'] = (funnel_abt['users_test'] / funnel_abt['users_test'].shift(1) * 100).round(1)

# Визуализируем step-by-step воронку

plt.figure(figsize=(10, 6))
plt.plot(funnel_abt['step'][2:], funnel_abt['conversion_from_previous_%_control'][2:], color='blue', label = 'control')
plt.plot(funnel_abt['step'][2:], funnel_abt['conversion_from_previous_%_test'][2:], color='red', label = 'test')
plt.title('Step-by-step воронка: конверсия от предыдущего шага')
plt.ylabel('Конверсия от предыдущего шага (%)')
plt.xlabel('Этапы воронки')
plt.xticks(rotation = 45)
plt.grid(True)
plt.legend()
plt.show()

# Выводим таблицу воронки

print(funnel_abt)
```

После построения **классической** воронки можно увидеть, что конверсия шагов от первого шага для контрольной группы изменяется стабильно с этапа установки `install / open_web` (100%) до завершения онбординга `onboarding_complete` (83.8%), после чего происходит резкое снижение на 17% на шаге совершения первого депозита `first_deposit` (66.6%) и ещё более резкое снижение на 30% и 23% на этапах приобретения первого актива `asset_purchase` (36.6%) и совершения второго депозита `second_deposit` (13.5%)

Тестовая группа показала себя хуже на этапах перехода к завершению онбординга `onboarding_complete` (75.8%) - на 8% меньше по сравнению с контрольной группой и совершения первого депозита `first_deposit` (65.4%) - на 1% меньше по сравнению с контрольной группой. Однако на этапах регистрации `install / open_web` (93.4%), приобретения первого актива `asset_purchase` (43.4%) и совершения второго депозита `second_deposit` (21.3%) тестовая группа показала наибольшие результаты - увеличение показателя конверсии на 0.3% для этапа регистрации и на 7% на этапах приобретения первого открытия и совершения второго депозита по сравнению с контрольной группой.

На этапе перехода к главной странице `main_page` показатель конверсии у контрольной и тестовой групп - одинаковый (85.8%)

После построения **step-by-step** воронки мы обратили внимание, что контрольная группа проявляет себя лучше лишь на этапах перехода на главную страницу `main_page`(92.2%) против 91.9% в тестовой группе и на этапе завершения онбординга `onboarding_complete` (97.6%) против 88.3% в тестовой группе. Но на всех остальных этапах тестовая группа показала лучшие результаты:

- регистрация `registration` (93.4%) - на 0.3% выше по сравнению с контрольной группой
- совершение первого депозита `first_deposit`(86.3%) - на 7% выше по сравнению с контрольной группой
- приобретение первого актива `asset_purchase`(66.2%) - на 12% выше по сравнению с контрольной группой
- совершение второго депозита `second_deposit`(49.3%) - на 13% выше по сравнению с контрольной группой

То есть, пользователи из тестовой группы лучше переходили на этапы совершения первого депозита, покупки первого актива и совершения второго депозита по сравнению с пользователями из контрольной группы

Таким образом, мы можем предполагать, что новая фича оказала положительное влияние на показатель конверсии, но на данном этапе анализа мы не можем сказать о том, значимы ли статистически данные изменения, поскольку статистический тест ещё не проведён

**Задача 3.3.** Влияние новой фичи на конверсию во второй депозит с учётом уровня риска купленного актива.

Для того, чтобы понять, как уровень риска первого актива влияет на конверсию во второй депозит, мы построим и для сравнения визуализируем (с помощью линейных графиков) **step-by-step** воронку относительно предыдущего шага отдельно для каждого уровня риска и каждой группы - контрольной и тестовой

```python
# Отбираем пользователей, которые приобрели актив

abt_asset_low_risk = abt[(abt['risk_level'] == 'low') & (abt['event_name'] == 'asset_purchase')]
abt_asset_medium_risk = abt[(abt['risk_level'] == 'medium') & (abt['event_name'] == 'asset_purchase')]
abt_asset_high_risk = abt[(abt['risk_level'] == 'high') & (abt['event_name'] == 'asset_purchase')]

# На основе предыдущих датафреймов отбираем пользователей, которые совершили второй депозит в зависимости от уровня риска 

abt_deposit2_low_risk = abt[(abt['event_name'] == 'second_deposit') & (abt['user_id'].isin(abt_asset_low_risk['user_id']))]
abt_deposit2_medium_risk = abt[(abt['event_name'] == 'second_deposit') & (abt['user_id'].isin(abt_asset_medium_risk['user_id']))]
abt_deposit2_high_risk = abt[(abt['event_name'] == 'second_deposit') & (abt['user_id'].isin(abt_asset_high_risk['user_id']))]

# Фильтруем датасеты, оставляя данные контрольной группы

сontrol_abt_asset_low_risk = abt_asset_low_risk[abt_asset_low_risk['group'] == 'control']
сontrol_abt_asset_medium_risk = abt_asset_medium_risk[abt_asset_medium_risk['group'] == 'control']
сontrol_abt_asset_high_risk = abt_asset_high_risk[abt_asset_high_risk['group'] == 'control']
сontrol_abt_deposit2_low_risk = abt_deposit2_low_risk[abt_deposit2_low_risk['group'] == 'control']
сontrol_abt_deposit2_medium_risk = abt_deposit2_medium_risk[abt_deposit2_medium_risk['group'] == 'control']
сontrol_abt_deposit2_high_risk = abt_deposit2_high_risk[abt_deposit2_high_risk['group'] == 'control']

# Фильтруем датасеты, оставляя данные тестовой группы

test_abt_asset_low_risk = abt_asset_low_risk[abt_asset_low_risk['group'] == 'test']
test_abt_asset_medium_risk = abt_asset_medium_risk[abt_asset_medium_risk['group'] == 'test']
test_abt_asset_high_risk = abt_asset_high_risk[abt_asset_high_risk['group'] == 'test']
test_abt_deposit2_low_risk = abt_deposit2_low_risk[abt_deposit2_low_risk['group'] == 'test']
test_abt_deposit2_medium_risk = abt_deposit2_medium_risk[abt_deposit2_medium_risk['group'] == 'test']
test_abt_deposit2_high_risk = abt_deposit2_high_risk[abt_deposit2_high_risk['group'] == 'test']

# Считаем количество пользователей контрольной группы

сontrol_abt_counts_asset_low_risk = сontrol_abt_asset_low_risk.groupby('event_name')['user_id'].nunique().reset_index()
сontrol_abt_counts_asset_medium_risk = сontrol_abt_asset_medium_risk.groupby('event_name')['user_id'].nunique().reset_index()
сontrol_abt_counts_asset_high_risk = сontrol_abt_asset_high_risk.groupby('event_name')['user_id'].nunique().reset_index()
сontrol_abt_counts_deposit2_low_risk = сontrol_abt_deposit2_low_risk.groupby('event_name')['user_id'].nunique().reset_index()
сontrol_abt_counts_deposit2_medium_risk = сontrol_abt_deposit2_medium_risk.groupby('event_name')['user_id'].nunique().reset_index()
сontrol_abt_counts_deposit2_high_risk = сontrol_abt_deposit2_high_risk.groupby('event_name')['user_id'].nunique().reset_index()

# Считаем количество пользователей тестовой группы

test_abt_counts_asset_low_risk = test_abt_asset_low_risk.groupby('event_name')['user_id'].nunique().reset_index()
test_abt_counts_asset_medium_risk = test_abt_asset_medium_risk.groupby('event_name')['user_id'].nunique().reset_index()
test_abt_counts_asset_high_risk = test_abt_asset_high_risk.groupby('event_name')['user_id'].nunique().reset_index()
test_abt_counts_deposit2_low_risk = test_abt_deposit2_low_risk.groupby('event_name')['user_id'].nunique().reset_index()
test_abt_counts_deposit2_medium_risk = test_abt_deposit2_medium_risk.groupby('event_name')['user_id'].nunique().reset_index()
test_abt_counts_deposit2_high_risk = test_abt_deposit2_high_risk.groupby('event_name')['user_id'].nunique().reset_index()

# Объединяем данные для контрольной группы

сontrol_abt_counts_low_risk = pd.concat([сontrol_abt_counts_asset_low_risk, сontrol_abt_counts_deposit2_low_risk], axis=0).reset_index()
сontrol_abt_counts_low_risk = сontrol_abt_counts_low_risk.drop('index', axis=1)
сontrol_abt_counts_medium_risk = pd.concat([сontrol_abt_counts_asset_medium_risk, сontrol_abt_counts_deposit2_medium_risk], axis=0).reset_index()
сontrol_abt_counts_medium_risk = сontrol_abt_counts_medium_risk.drop('index', axis=1)
сontrol_abt_counts_high_risk = pd.concat([сontrol_abt_counts_asset_high_risk, сontrol_abt_counts_deposit2_high_risk], axis=0).reset_index()
сontrol_abt_counts_high_risk = сontrol_abt_counts_high_risk.drop('index', axis=1)

# Объединяем данные для тестовой группы

test_abt_counts_low_risk = pd.concat([test_abt_counts_asset_low_risk, test_abt_counts_deposit2_low_risk], axis=0).reset_index()
test_abt_counts_low_risk = test_abt_counts_low_risk.drop('index', axis=1)
test_abt_counts_medium_risk = pd.concat([test_abt_counts_asset_medium_risk, test_abt_counts_deposit2_medium_risk], axis=0).reset_index()
test_abt_counts_medium_risk = test_abt_counts_medium_risk.drop('index', axis=1)
test_abt_counts_high_risk = pd.concat([test_abt_counts_asset_high_risk, test_abt_counts_deposit2_high_risk], axis=0).reset_index()
test_abt_counts_high_risk = test_abt_counts_high_risk.drop('index', axis=1)

# Считаем конверсию для контрольной группы

сontrol_abt_low_risk_funnel = сontrol_abt_counts_low_risk['user_id'] / сontrol_abt_counts_low_risk['user_id'].shift(1) * 100
сontrol_abt_low_risk_funnel = сontrol_abt_low_risk_funnel.fillna(100)
сontrol_abt_medium_risk_funnel = сontrol_abt_counts_medium_risk['user_id'] / сontrol_abt_counts_medium_risk['user_id'].shift(1) * 100
сontrol_abt_medium_risk_funnel = сontrol_abt_medium_risk_funnel.fillna(100)
сontrol_abt_high_risk_funnel = сontrol_abt_counts_high_risk['user_id'] / сontrol_abt_counts_high_risk['user_id'].shift(1) * 100
сontrol_abt_high_risk_funnel = сontrol_abt_high_risk_funnel.fillna(100)

# Считаем конверсию для тестовой группы

test_abt_low_risk_funnel = test_abt_counts_low_risk['user_id'] / test_abt_counts_low_risk['user_id'].shift(1) * 100
test_abt_low_risk_funnel = test_abt_low_risk_funnel.fillna(100)
test_abt_medium_risk_funnel = test_abt_counts_medium_risk['user_id'] / test_abt_counts_medium_risk['user_id'].shift(1) * 100
test_abt_medium_risk_funnel = test_abt_medium_risk_funnel.fillna(100)
test_abt_high_risk_funnel = test_abt_counts_high_risk['user_id'] / test_abt_counts_high_risk['user_id'].shift(1) * 100
test_abt_high_risk_funnel = test_abt_high_risk_funnel.fillna(100)

# Соберём через словарь данные в датафрейм

funnel_compare3 = pd.DataFrame({
    'step': funnel_steps[6:],
    'control_low_risk': сontrol_abt_low_risk_funnel.round(1).values,
    'control_medium_risk': сontrol_abt_medium_risk_funnel.round(1).values,
    'control_high_risk': сontrol_abt_high_risk_funnel.round(1).values,
    'test_low_risk': test_abt_low_risk_funnel.round(1).values,
    'test_medium_risk': test_abt_medium_risk_funnel.round(1).values,
    'test_high_risk': test_abt_high_risk_funnel.round(1).values
})

# Визуализируем воронку для low_risk

plt.figure(figsize=(10, 6))
plt.plot(funnel_compare3['step'], funnel_compare3['control_low_risk'], color='blue', label = 'control')
plt.plot(funnel_compare3['step'], funnel_compare3['test_low_risk'], color='red', label = 'test')
plt.title('Step-by-step воронка для low_risk: конверсия от предыдущего шага')
plt.ylabel('Конверсия от первого шага (%)')
plt.xlabel('Этапы воронки')
plt.grid(True)
plt.legend()
plt.show()

# Визуализируем воронку для medium_risk

plt.figure(figsize=(10, 6))
plt.plot(funnel_compare3['step'], funnel_compare3['control_medium_risk'], color='blue', label = 'control')
plt.plot(funnel_compare3['step'], funnel_compare3['test_medium_risk'], color='red', label = 'test')
plt.title('Step-by-step воронка для medium_risk: конверсия от предыдущего шага')
plt.ylabel('Конверсия от первого шага (%)')
plt.xlabel('Этапы воронки')
plt.grid(True)
plt.legend()
plt.show()

# Визуализируем воронку для high_risk

plt.figure(figsize=(10, 6))
plt.plot(funnel_compare3['step'], funnel_compare3['control_high_risk'], color='blue', label = 'control')
plt.plot(funnel_compare3['step'], funnel_compare3['test_high_risk'], color='red', label = 'test')
plt.title('Step-by-step воронка для high_risk: конверсия от предыдущего шага')
plt.ylabel('Конверсия от первого шага (%)')
plt.xlabel('Этапы воронки')
plt.grid(True)
plt.legend()
plt.show()

# Выводим таблицу воронки

print(funnel_compare3)
```

После построения **step-by-step** воронки относительно предыдущего шага отдельно для каждого уровня риска, мы видим, что тенденция "с увеличением уровня риска снижается конверсия из этапа приобретения первого актива в этап совершения второго депозита" осталась прежней по сравнению с анализом исторических данных

Однако важно отметить то, что тестовая группа проявила себя лучше по сравнению с контрольной - показатели конверсии из этапа приобретения первого актива в этап совершения второго депозита увеличились:

- Конверсия для активов с низким риском увеличилась на **7%**
- Конверсия для активов со средним риском увеличилась на **2,5%**
- Конверсия для активов с высоким риском увеличилась на **19%**

Возможная причина большого увеличения конверсии для активов с высоким риском - влияние новой фичи, то есть, обновление онбординга

Рассчитаем, является ли изменение в конверсии перехода из этапа приобретения первого актива с разными рисками в этап открытия второго депозита статистически значимым. Для этого используем статистический `Z-тест пропорций`, поскольку диапазон значений нашей метрики колеблется от 0 до 1. 

Сформулируем три гипотезы:

1 гипотеза:

- **H0:** p_a = p_b - конверсия из этапа приобретения первого актива с низким риском в этап совершения второго депозита для пользователей контрольной и тестовой группы - равна

- **H1:** p_a < p_b - конверсия из этапа приобретения первого актива с низким риском в этап совершения второго депозита для пользователей тестовой группы выше, чем для пользователей контрольной группы и это изменение - стастически значимо

2 гипотеза:

- **H0:** p_a = p_b - конверсия из этапа приобретения первого актива со средним риском в этап совершения второго депозита для пользователей контрольной и тестовой группы - равна

- **H1:** p_a < p_b - конверсия из этапа приобретения первого актива со средним риском в этап совершения второго депозита для пользователей тестовой группы выше, чем для пользователей контрольной группы и это изменение - стастически значимо

3 гипотеза:

- **H0:** p_a = p_b - конверсия из этапа приобретения первого актива с высоким риском в этап совершения второго депозита для пользователей контрольной и тестовой группы - равна

- **H1:** p_a < p_b - конверсия из этапа приобретения первого актива с высоким риском в этап совершения второго депозита для пользователей тестовой группы выше, чем для пользователей контрольной группы и это изменение - стастически значимо

Начнём с проверки гипотезы, связанной с сегментом пользователей, приобретавших активы с низким риском

```python
# Проверка гипотезы, связанной с сегментом пользователей, приобретавших активы с низким риском

n_a = сontrol_abt_counts_low_risk['user_id'].loc[0] # количество купленных активов в контрольной группе

n_b = test_abt_counts_low_risk['user_id'].loc[0] # количество купленных активов в тестовой группе

m_a = сontrol_abt_counts_low_risk['user_id'].loc[1] # количество совершенных вторых депозитов в контрольной группе

m_b = test_abt_counts_low_risk['user_id'].loc[1] # количество совершенных вторых депозитов в тестовой группе

p_a, p_b = m_a/n_a, m_b/n_b # доли совершенных вторых депозитов для каждой группы: контрольной и тестовой

if (p_a*n_a>10)and((1-p_a)*n_a>10)and(p_b*n_b>10)and((1-p_b)*n_b>10): # проверяем предпосылку о достаточном количестве данных
    print('Предпосылка о достаточном количестве данных выполняется!')
else:
    print('Предпосылка о достаточном количестве данных не выполняется!')

alpha = 0.05 # уровень значимости


stat_ztest, p_value_ztest = proportions_ztest(
  [m_a, m_b],
  [n_a, n_b],
  alternative='smaller' # проводим Z-тест пропорций, H0: p_a = p_b, H1: p_a < p_b, так как мы проверяем: 
                        # значимо ли статистически увеличение конверсии из этапа приобретения первого актива с низким риском 
                        # в этап совершения второго депозита в тестовой группе по сравнению с контрольной группой
) 

if p_value_ztest > alpha:
    print(f'pvalue={p_value_ztest} > {alpha}')
    print('Не получилось отвергнуть нулевую гипотезу')
    print('Статистически значимое увеличение конверсии из этапа приобретения первого актива с низким риском в этап совершения второго депозита в тестовой группе по сравнению с контрольной группой - не наблюдается')
else:
    print(f'pvalue={p_value_ztest} < {alpha}')
    print('Отвергаем нулевую гипотезу')
    print('Наблюдается статистически значимое увеличение конверсии из этапа приобретения первого актива с низким риском в этап совершения второго депозита в тестовой группе по сравнению с контрольной группой')
```

Начнём с проверки гипотезы, связанной с ключевой метрикой

```python
# Получаем список всех уникальных групп

groups = sorted(abt['group'].unique())

# Группируем данные

deposit_per_user = abt.groupby(['user_id', 'group'])['amount'].sum().reset_index()

# Создаём словарь для хранения данных по группам

group_data = {group: deposit_per_user[deposit_per_user['group'] == group]['amount'] for group in groups}

# Создаём список для хранения результатов

results1 = []

# Проводим попарные сравнения всех групп

for i in range(len(groups)):
   for j in range(i + 1, len(groups)):
       group1 = groups[i]
       group2 = groups[j]
       _, p_value = ttest_ind(group_data[group1], group_data[group2]) 

       mean_a = group_data[group1].mean() # рассчитываем ключевую метрику
       mean_b = group_data[group2].mean() # рассчитываем ключевую метрику
       absolute_lift = mean_b - mean_a
       relative_lift = (group_data[group2].mean() / group_data[group1].mean() - 1) * 100

       # Добавляем результаты в список
       results1.append({
           'Group A': group1,
           'Group B': group2,
           'p-value': p_value,
           'Metric A': round(mean_a, 4),
           'Metric B': round(mean_b, 4),
           'Absolute Lift': round(absolute_lift, 4),
           'Relative Lift (%)': round(relative_lift, 4)
})

# Создаём DataFrame из результатов

results_1 = pd.DataFrame(results1)

# Выводим результат

print('Ключевая метрика — средняя сумма всех депозитов на одного пользователя \n(включая тех, кто установил приложение или открыл веб-версию):')
display(results_1)
```

```python
# Вычисляем необходимые переменные

control_reg = abt[(abt['group'] == 'control') & (abt['event_name'] == 'registration')]['user_id'].nunique()
test_reg = abt[(abt['group'] == 'test') & (abt['event_name'] == 'registration')]['user_id'].nunique()

control_dep1 = abt[(abt['group'] == 'control') & (abt['event_name'] == 'first_deposit')]['user_id'].nunique()
test_dep1 = abt[(abt['group'] == 'test') & (abt['event_name'] == 'first_deposit')]['user_id'].nunique()

# Проверка гипотезы, связанной с барьерной метрикой

n_a = control_reg # количество регистраций в контрольной группе

n_b = test_reg # количество регистраций в тестовой группе

m_a = control_dep1 # количество совершенных первых депозитов в контрольной группе

m_b = test_dep1 # количество совершенных первых депозитов в тестовой группе

p_a, p_b = m_a/n_a, m_b/n_b # доли совершенных первых депозитов для каждой группы: контрольной и тестовой

if (p_a*n_a>10)and((1-p_a)*n_a>10)and(p_b*n_b>10)and((1-p_b)*n_b>10): # проверяем предпосылку о достаточном количестве данных
    print('Предпосылка о достаточном количестве данных выполняется!\n')
else:
    print('Предпосылка о достаточном количестве данных не выполняется!\n')

alpha = 0.05 # уровень значимости

stat_ztest, p_value_ztest = proportions_ztest(
  [m_a, m_b],
  [n_a, n_b],
  alternative='smaller' # проводим Z-тест пропорций, H0: p_a = p_b, H1: p_a < p_b, так как мы проверяем: 
                        # значимо ли статистически увеличение конверсии из этапа регистрации в этап 
                        # совершения первого депозита в тестовой группе по сравнению с контрольной группой
) 

if p_value_ztest > alpha:
    print(f'pvalue={p_value_ztest} > {alpha}')
    print('Не получилось отвергнуть нулевую гипотезу')
    print('Статистически значимое увеличение конверсии из этапа регистрации в этап совершения первого депозита в тестовой группе по сравнению с контрольной группой - не наблюдается')
else:
    print(f'pvalue={p_value_ztest} < {alpha}')
    print('Отвергаем нулевую гипотезу')
    print('Наблюдается статистически значимое увеличение конверсии из этапа регистрации в этап совершения первого депозита в тестовой группе по сравнению с контрольной группой')

# Рассчитываем абсолютные и относительные изменения значений в тестовой группе по сравнению с контрольной     
    
absolute_lift = p_b - p_a
relative_lift = (p_b / p_a - 1) * 100

# Соберём через словарь данные в датафрейм

results_2 = pd.DataFrame({
    'Group A': 'control',
    'Group B': 'test',
    'p-value': [p_value_ztest],
    'Metric A': [round(p_a, 4)],
    'Metric B': [round(p_b, 4)],
    'Absolute Lift': [round(absolute_lift, 4)],
    'Relative Lift (%)': [round(relative_lift, 4)]
})

# Выводим результат

print('\nБарьерная метрика — конверсия из регистрации в первый депозит:')
display(results_2)
```

Так как `pvalue = 0.9397282759544358 > alpha = 0.05`, мы можем сделать вывод о том, что мы не можем отвергнуть нулевую гипотезу. Это значит, что статичестически значимое увеличение барьерной метрики (конверсии из этапа регистрации в этап совершения первого депозита) в тестовой группе по сравнению с контрольной группой при уровне значимости `alpha = 0.05` - не наблюдается. - накопленную динамику изменения по дням эксперимента для каждой группы,
- стабильность p-value во время эксперимента.

```python
# Формируем лист дат для цикла и сортируем его

date_list = abt['event_ts'].dt.date.unique()
date_list.sort()

# Пустой список для результатов

records = []

# Расчет кумулятивной метрики

for current_date in date_list:

    # Все события до текущей даты
    
    active_to_date = abt[abt['event_ts'].dt.date <= current_date]

    # Сумма всех депозитов на одного пользователя
    
    deposit_per_user = active_to_date.groupby(['user_id', 'group'])['amount'].sum().reset_index()

    # Средняя сумма всех депозитов на одного пользователя
    
    avg_deposit_per_user = deposit_per_user.groupby('group')['amount'].mean().reset_index()
   
    avg_deposit_per_user['date'] = current_date
    records.append(avg_deposit_per_user)

# Собираем датафрейм

avg_deposit_per_user_df = pd.concat(records)

# Построим график

plt.figure(figsize=(10, 5))

# Цикл по группам

for group in avg_deposit_per_user_df['group'].unique():
    
    # Фильтрация данных по группе
    
    group_data = avg_deposit_per_user_df[avg_deposit_per_user_df['group'] == group]
    
    # Сортировка по дате для правильного отображения линии
    
    group_data = group_data.sort_values('date')
    
    # Построение линии для группы
    
    plt.plot(group_data['date'], group_data['amount'], label=group, marker='o')

plt.title('Накопленная динамика изменения ключевой метрики по дням эксперимента для каждой группы')
plt.xlabel('Дата')
plt.ylabel('Средняя сумма всех депозитов на одного пользователя')
plt.grid(True)
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Расчёт p-value по дням

# Пустой список для результатов

pval_records = []

for current_date in date_list:
    
    # Все события до текущей даты
    
    active_to_date = abt[abt['event_ts'].dt.date <= current_date]
    
    # Сумма всех депозитов на одного пользователя
    
    deposit_per_user = active_to_date.groupby(['user_id', 'group'])['amount'].sum().reset_index()

    # Объявление переменных с данными активности по каждой группе
    
    control = deposit_per_user[deposit_per_user['group'] == 'control']['amount']
    test = deposit_per_user[deposit_per_user['group'] == 'test']['amount']

    #расчет p-value
    
    if len(control) > 1 and len(test) > 1:
        p_val_ab = ttest_ind(control, test, equal_var=False, nan_policy='omit').pvalue
    else:
        p_val_ab = np.nan

    pval_records.append({
        'date': current_date,
        'p_value_ab': p_val_ab
    })


# Собираем в DataFrame

pvalue_df = pd.DataFrame(pval_records)

# График

plt.figure(figsize=(10, 5))

# Построение линий графика

plt.plot(pvalue_df['date'], pvalue_df['p_value_ab'], label='Control vs Test', color = 'purple')

plt.axhline(y=0.05, linestyle='--', color='black', label='alpha') 
plt.title('Динамика p-value по дням эксперимента')
plt.xlabel('Дата')
plt.ylabel('p-value')
plt.xticks(rotation=45)
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

Исходя из полученных графиков мы можем сделать следующие выводы:

- График `Накопленная динамика изменения ключевой метрики по дням эксперимента для каждой группы`:
    - Разница между контрольной и тестовыми группами растёт по мере продвижения эксперимента. Это может говорить о накопительном эффекте от изменений: пользователи становятся более активными, изменения в тестовой группе потенциально положительно повлияли на среднюю сумму всех депозитов на одного пользователя.
    - На протяжении большей части эксперимента контрольная группа опережала тестовую, однако в последние дни показатели ключевой метрики тестовой группы превысили показатели ключевой метрики контрольной группы
    
- График `Динамика p-value по дням эксперимента`:
    - На первых днях эксперимента p-value был ниже порога значимости, однако менее чем через неделю после начала эксперимента, p-value превысил порог значимости и не опускался ниже него до конца эксперимента

Таким образом, эксперимент нельзя считать успешным, поскольку для ключевой и барьерной метрик статистически значимых изменений - не наблюдается. При этом изменения двух вспомогательных метрик являются статистически значимыми, что означает, что фича нуждатся в более качественной доработке и обновлении. ### 4. Анализ изменений суммы депозитов на платящего пользователя

Новая фича могла повлиять на поведение пользователей.
- Пользователи, которые раньше вносили небольшие суммы, могли стать более осторожными, сократить свои вложения или совсем перестать платить. Это отразится в снижении 25-го перцентиля суммы депозитов в тестовой группе.
- Пользователи, которые склонны к более крупным инвестициям, могли сильнее вовлечься в продукт и начать вносить больше средств. Это отразится в росте 75-го перцентиля.

Используя бутстрап, сравним разницы перцентилей (25, 50, 75) суммы всех депозитов на платящего пользователя в контрольной и тестовой группах:
- Построим доверительные интервалы для разницы между группами для каждого перцентиля.
- Интерпретируем изменения в контексте сформулированной выше гипотезы.
- Визуализируем результат бутстрап-теста. 

```python
# Фильтруем датасет, оставляя данные контрольной группы с платящими пользователями

control_dep1 = abt[(abt['group'] == 'control') & (abt['amount'] > 0)]['amount']

# Фильтруем датасет, оставляя данные тестовой группы с платящими пользователями

test_dep1 = abt[(abt['group'] == 'test') & (abt['amount'] > 0)]['amount']

# Задаём количество итераций и создаём пустой список для сохранения результатов

n_iterations = 1000
boot_diffs_25 = []
boot_diffs_50 = []
boot_diffs_75 = []

# Бутстрап-разница 25, 50 и 75 перцентилей между тестовой и контрольной группами

for i in range(n_iterations):
    boot_control = np.random.choice(control_dep1, size=len(control), replace=True)
    boot_test = np.random.choice(test_dep1, size=len(test), replace=True)

    # Для расчёта 25 перцентиля используем .percentile() из библиотеки numpy
    
    control_25p = np.percentile(boot_control, 25)
    test_25p = np.percentile(boot_test, 25)

    # Находим разницу между значениями тестовой и контрольной групп
    
    boot_diffs_25.append(test_25p - control_25p)
    
    # Для расчёта 95 перцентиля используем .percentile() из библиотеки numpy 
    
    control_50p = np.percentile(boot_control, 50)
    test_50p = np.percentile(boot_test, 50)

    # Находим разницу между значениями тестовой и контрольной групп
    
    boot_diffs_50.append(test_50p - control_50p)
    
    # Для расчёта 95 перцентиля используем .percentile() из библиотеки numpy 
    
    control_75p = np.percentile(boot_control, 75)
    test_75p = np.percentile(boot_test, 75)

    # Находим разницу между значениями тестовой и контрольной групп
    
    boot_diffs_75.append(test_75p - control_75p)

# Создаём массив разниц между 95-ми перцентилями тестовой и контрольной групп

boot_diffs_25 = np.array(boot_diffs_25)
boot_diffs_50 = np.array(boot_diffs_50)
boot_diffs_75 = np.array(boot_diffs_75)

# Доверительный интервал разницы между 95-ми перцентилями (перцентильный метод)

diff_ci_25 = np.percentile(boot_diffs_25, [2.5, 97.5])
diff_ci_50 = np.percentile(boot_diffs_50, [2.5, 97.5])
diff_ci_75 = np.percentile(boot_diffs_75, [2.5, 97.5])

# Визуализируем доверительный интервал для 25-ого перцинтиля

plt.figure(figsize=(12, 8))

plt.hist(boot_diffs_25, bins=30, color='skyblue', edgecolor='black')
plt.axvline(diff_ci_25[0], color='red', linestyle='--', label='2.5-й перцентиль')
plt.axvline(diff_ci_25[1], color='red', linestyle='--', label='97.5-й перцентиль')
plt.title("Бутстрап-распределение: разность 25-перцентилей")
plt.legend()
plt.show()

# Вывод доверительного интервала для 25-ого перцинтиля

print(f"25-й перцентиль контрольной группы: {np.percentile(control_dep1, 25):.2f}")
print(f"25-й перцентиль тестовой группы: {np.percentile(test_dep1, 25):.2f}")
print(f"95%-й доверительный интервал разности 25-х перцентилей [{diff_ci_25[0]:.2f}, {diff_ci_25[1]:.2f}]")

# Визуализируем доверительный интервал для 50-ого перцинтиля

plt.figure(figsize=(12, 8))

plt.hist(boot_diffs_50, bins=30, color='skyblue', edgecolor='black')
plt.axvline(diff_ci_50[0], color='red', linestyle='--', label='2.5-й перцентиль')
plt.axvline(diff_ci_50[1], color='red', linestyle='--', label='97.5-й перцентиль')
plt.title("Бутстрап-распределение: разность 50-перцентилей")
plt.legend()
plt.show()

# Вывод доверительного интервала для 50-ого перцинтиля

print(f"50-й перцентиль контрольной группы: {np.percentile(control_dep1, 50):.2f}")
print(f"50-й перцентиль тестовой группы: {np.percentile(test_dep1, 50):.2f}")
print(f"95%-й доверительный интервал разности 50-х перцентилей [{diff_ci_50[0]:.2f}, {diff_ci_50[1]:.2f}]")

# Визуализируем доверительный интервал для 75-ого перцинтиля

plt.figure(figsize=(12, 8))

plt.hist(boot_diffs_75, bins=30, color='skyblue', edgecolor='black')
plt.axvline(diff_ci_75[0], color='red', linestyle='--', label='2.5-й перцентиль')
plt.axvline(diff_ci_75[1], color='red', linestyle='--', label='97.5-й перцентиль')
plt.title("Бутстрап-распределение: разность 75-перцентилей")
plt.legend()
plt.show()

# Вывод доверительного интервала для 75-ого перцинтиля

print(f"75-й перцентиль контрольной группы: {np.percentile(control_dep1, 75):.2f}")
print(f"75-й перцентиль тестовой группы: {np.percentile(test_dep1, 75):.2f}")
print(f"95%-й доверительный интервал разности 75-х перцентилей [{diff_ci_75[0]:.2f}, {diff_ci_75[1]:.2f}]")
```
